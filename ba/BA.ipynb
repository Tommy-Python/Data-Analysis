## 1.3 How often do users come back?

```python
firstvisit = visits.groupby('Uid')["month Start Ts"].min().reset_index()
firstvisit.info()
```

```python
firstvisit
```

```python


firstvisit.rename(columns={'month Start Ts':'first_visit_month'}, inplace=True)
firstvisit.info()
```

```python
visits = visits.merge(firstvisit,on='Uid')
```

```python
visits.head()
```

```python
visits.rename(columns={'Uid':'player_id'}, inplace=True)
visits.rename(columns={'Start Ts':'purchase_datetime'}, inplace=True)
visits["purchase_id"] = visits["player_id"]
purchases = visits

first_purchase_dates = purchases.groupby('player_id')['purchase_datetime'].min()
first_purchase_dates.name = 'first_purchase_datetime'
purchases = purchases.join(first_purchase_dates, on='player_id')
purchases['purchase_month'] = purchases['purchase_datetime'].astype(
    'datetime64[M]'
)
purchases['first_purchase_month'] = purchases[
    'first_purchase_datetime'
].astype('datetime64[M]')
purchases_grouped_by_cohorts = purchases.groupby(
    ['first_purchase_month', 'purchase_month']
).agg({'player_id': ["count",'nunique']}).reset_index()
purchases_grouped_by_cohorts.rename(columns={0:'first_purchase'}, inplace=True)
purchases_grouped_by_cohorts.columns = ['first_purchase_month', 'purchase_month',"all_visitor", "unique_visitor"]
```

```python

purchases_grouped_by_cohorts['purchases_per_player'] = (
    purchases_grouped_by_cohorts['all_visitor']
    / purchases_grouped_by_cohorts['unique_visitor']
)

purchases_grouped_by_cohorts['cohort_lifetime'] = (
    purchases_grouped_by_cohorts['purchase_month']
    - purchases_grouped_by_cohorts['first_purchase_month']
)
purchases_grouped_by_cohorts['cohort_lifetime'] = purchases_grouped_by_cohorts[
    'cohort_lifetime'
] / np.timedelta64(1, 'M')
purchases_grouped_by_cohorts['cohort_lifetime'] = (
    purchases_grouped_by_cohorts['cohort_lifetime'].round().astype('int')
)
lifetime_pivot = purchases_grouped_by_cohorts.pivot_table(
    index='first_purchase_month',
    columns='cohort_lifetime',
    values='purchases_per_player',
    aggfunc='mean',
) 
print (lifetime_pivot) 
```

```python
lifetime_pivot.index=lifetime_pivot.index.strftime('%B, %Y')
```

```python
plt.figure(figsize=(13,9))
plt.title("ComebackUser")

sns.heatmap(
    lifetime_pivot,
    cmap="plasma",
    annot=True,
    fmt='.2f',
    linewidths=1,
    linecolor='black')
plt.show()
```

```python

initial_users_count = purchases_grouped_by_cohorts[purchases_grouped_by_cohorts['cohort_lifetime'] == 0][['first_purchase_month','unique_visitor']]

initial_users_count = initial_users_count.rename(columns={'unique_visitor':'cohort_users'})

purchases_grouped_by_cohorts = purchases_grouped_by_cohorts.merge(initial_users_count, on='first_purchase_month')

purchases_grouped_by_cohorts['retention'] =  purchases_grouped_by_cohorts['unique_visitor']/purchases_grouped_by_cohorts['cohort_users'] 

retention_pivot = purchases_grouped_by_cohorts.pivot_table(
    index='first_purchase_month',
    columns='cohort_lifetime',
    values='retention',
    aggfunc='sum')

print(retention_pivot)
```

```python
retention_pivot.index=retention_pivot.index.strftime('%B, %Y')
```

```python

plt.figure(figsize=(13,9))
plt.title("Retention")

sns.heatmap(
    retention_pivot,
    cmap="plasma",
    annot=True,
    fmt='.2%',
    vmax=0.1,
    linewidths=1,
    linecolor='black')
plt.show()
```

## 1.3 Conclusion

In the Graphs we can see, between 4.85 % and 8.55% came back. The September 2017 performes best with 8,55 % cameback while the durability of the june 2017 is oustanding 4.5 % over the entire year cameback.

```python
orders_visits=orders_visits.dropna()
```

```python
orders_visits
```

```python
orders_visits["visitday_to_buyday"]=(orders_visits["Buy Ts"] - orders_visits["date Start Ts"]) / np.timedelta64(1, 'D')
```

```python
def getdays(row):
    
    if row["visitday_to_buyday"] <1:
        x="0"
    elif row["visitday_to_buyday"] ==1:
        x="1"
    elif row["visitday_to_buyday"] <7:
        x="7"
    elif row["visitday_to_buyday"] <14:
        x="14"
    elif row["visitday_to_buyday"] <30:
        x="30"
    else:
        x=">30"
    return x
orders_visits['visitday_to_buyday'] = orders_visits.apply(getdays,axis=1)
```

```python
orders_visits.head(10)
```

```python
orders_visits["month"]= orders_visits["date Start Ts"].astype('datetime64[M]')
buy_cohort_pivot = orders_visits.pivot_table(
    index='month',
    columns='visitday_to_buyday',
    values='Uid',
    aggfunc='count')
```
